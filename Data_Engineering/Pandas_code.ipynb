# 1. Importing Pandas

```python
import pandas as pd
```

Explanation:
Loads the Pandas library so we can work with DataFrames.

# 2. Creating a DataFrame

```python
data = {
    "Name": ["Ali", "Sara", "John"],
    "Age": [25, 30, 28],
    "City": ["Delhi", "Hyderabad", "Mumbai"]
}

df = pd.DataFrame(data)
df
```

# 3.Basic Data Inspection

```python
df.head()        # first 5 rows
df.tail()        # last 5 rows
df.info()        # column types + summary
df.describe()    # statistics
```

# 4.Selecting Columns

```python
df["Name"]
df[["Name", "City"]]
```

# 5.Filtering Rows

```python
df[df["Age"] > 25]
```

# 7. Adding a New Column

```python
df["Age_in_5_years"] = df["Age"] + 5
df
```

# 8. Handling Missing Values

```python
df.dropna()             # remove rows with nulls
df.fillna("Unknown")    # replace nulls with a value
```

# âœ… 9. Sorting Data

```python
df.sort_values(by="Age", ascending=False)
```

# 10. Merging Two DataFrames

```python
orders = pd.DataFrame({
    "OrderID": [1, 2],
    "CustomerID": [101, 102]
})

customers = pd.DataFrame({
    "CustomerID": [101, 102],
    "Name": ["Ali", "Sara"]
})

merged = pd.merge(orders, customers, on="CustomerID")
merged
```

# 11. Grouping & Aggregation

```python
df.groupby("City")["Age"].mean()
```
