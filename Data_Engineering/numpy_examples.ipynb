```python
import numpy as np

# 1. Creating arrays
arr1 = np.array([1, 2, 3])
print("1D Array:", arr1)

arr2 = np.array([[1, 2], [3, 4]])
print("2D Array:\n", arr2)

# 2. Arrays with zeros, ones, random values
zeros_arr = np.zeros((2, 3))
ones_arr = np.ones((2, 3))
rand_arr = np.random.rand(2, 3)

print("Zeros Array:\n", zeros_arr)
print("Ones Array:\n", ones_arr)
print("Random Array:\n", rand_arr)

# 3. Basic operations
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print("Addition:", a + b)
print("Multiplication:", a * b)
print("Square Root:", np.sqrt(a))

# 4. Broadcasting
print("Broadcasting Example:", a + 5)

# 5. Statistical operations
print("Mean:", np.mean(a))
print("Sum:", np.sum(a))

# 6. Dot product (matrix multiplication)
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])
print("Dot Product:\n", np.dot(m1, m2))

# 7. Slicing
slice_example = a[0:2]
print("Slice:", slice_example)
```
