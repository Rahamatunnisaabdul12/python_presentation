```python
import boto3
import logging

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

def create_vpn_infra(region="ap-south-1"):  # You can change the region as needed
    """Create a VPC and VPN Gateway."""
    ec2 = boto3.client('ec2', region_name=region)

    try:
        # Create a VPC
        vpc = ec2.create_vpc(CidrBlock='10.0.0.0/16')
        vpc_id = vpc['Vpc']['VpcId']
        logging.info(f"‚úÖ VPC created: {vpc_id}")

        # Create a VPN Gateway
        vpn_gateway = ec2.create_vpn_gateway(Type='ipsec.1')
        vpn_gateway_id = vpn_gateway['VpnGateway']['VpnGatewayId']
        logging.info(f"‚úÖ VPN Gateway created: {vpn_gateway_id}")

        # Attach VPN Gateway to VPC
        ec2.attach_vpn_gateway(VpcId=vpc_id, VpnGatewayId=vpn_gateway_id)
        logging.info("üîó VPN Gateway attached to VPC successfully.")

    except Exception as e:
        logging.error(f"‚ùå Error creating VPN infrastructure: {e}")

if __name__ == "__main__":
    create_vpn_infra(region="ap-south-1")  # Replace with your AWS region
```
