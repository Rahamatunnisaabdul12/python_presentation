```python
"""
Example 1: Basic Data Lens Registration
---------------------------------------
Simulates registering a dataset with logical and physical definitions.
"""

import json
import logging

# Setup logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

def register_dataset(logical, physical):
    """Simulate registering a dataset in the catalog."""
    dataset_entry = {"logical": logical, "physical": physical}
    with open("dataset_registration.json", "w") as f:
        json.dump(dataset_entry, f, indent=4)
    logging.info("âœ… Dataset registered successfully.")
    return dataset_entry

if __name__ == "__main__":
    logical_dataset = {
        "logical_name": "Customer Orders",
        "business_owner": "Sales Analytics",
        "description": "Processed customer order data for analysis."
    }

    physical_dataset = {
        "storage_path": "s3://datalake/trusted/orders/2025/",
        "schema": ["order_id", "customer_id", "order_date", "amount"],
        "refresh_frequency": "daily"
    }

    register_dataset(logical_dataset, physical_dataset)
```
